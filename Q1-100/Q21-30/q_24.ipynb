**24. Lexicographic Permutations**

image.png

```python
import math

place = 1_000_000
initial_digits = list(range(10))
results = []

for i in range(9, -1, -1):
  count = 0
  perm_value = math.perm(i, i)

  while (place - perm_value) > 0:
    count += 1
    place -= perm_value

  selected_digit = initial_digits.pop(count)
  results.append(selected_digit)

print(''.join(map(str, results)))
```
