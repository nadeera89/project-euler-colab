```python
import base64
import requests

url = 'https://api.github.com/repos/nadeera89/project-euler-colab/contents/Q1-100/Q21-30/q22_names.txt'
req = requests.get(url)
if req.status_code == requests.codes.ok:
    req = req.json()
    content = base64.b64decode(req['content']).decode("utf-8").replace('"','').upper().split(",")
    content.sort()
    print(content)
else:
    print('Content was not found.')
```

```python
alphabet_values = {chr(i): i - 64 for i in range(65, 91)}
print(alphabet_values)
```

```python
def get_name_value(word):
  return sum([alphabet_values[letter] for letter in word])
```

```python
total_name_score = 0

for name in content:
  total_name_score += (content.index(name) + 1) * get_name_value(name)

print(total_name_score)
```
