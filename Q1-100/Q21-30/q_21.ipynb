**21. Amicable Numbers**

image.png

```python
def find_all_divisors(n):
  sum = 1

  for i in range(2, int(n**0.5) + 1):
    if n % i == 0:
      sum += i
      if i * i != n:
        sum += (n // i)

  return sum
```

```python
amic_sum = [0] * 10_000
total = 0

for i in range(2, 10_000):
  if amic_sum[i] == 0:
    amic_sum[i] = find_all_divisors(i)
  if amic_sum[i] > i and amic_sum[i] < 10_000:
    amic_sum[amic_sum[i]] = find_all_divisors(amic_sum[i])
    if i == amic_sum[amic_sum[i]]:
      print(">>", amic_sum[i], i)
      total += (i + amic_sum[i])

print(total)
```
