**25. 1000-digit Fibonacci Number**

image.png

```python
prev = [0] * 1_000
curr = [0] * 1_000
next = [0] * 1_000

prev[0] = curr[0] = 1
```

```python
fib_index = 2

while next[999] == 0:
  fib_index += 1

  remainder = 0
  for i in range(1_000):
    sum_digit = (prev[i] + curr[i] + remainder)

    next[i] = sum_digit % 10
    remainder = sum_digit // 10

    prev[i] = curr[i]
    curr[i] = next[i]

fib_index
```
