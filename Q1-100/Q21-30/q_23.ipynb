**23. Non-Abundant Sums**

image.png

```python
def is_abundant_number(n):
  sum_of_divisors = 1

  for i in range(2, int(n**0.5) + 1):
    if n % i == 0:
      sum_of_divisors += i
      if i * i != n:
        sum_of_divisors += (n // i)

  return sum_of_divisors > n
```

```python
list_abundant_nums = []

for i in range(12, 28124):
  if is_abundant_number(i):
    list_abundant_nums.append(i)
```

```python
num_list = [False] * 28_123

for i in range(len(list_abundant_nums)):
  for j in range(i, len(list_abundant_nums)):
    value = list_abundant_nums[i] + list_abundant_nums[j]

    if value > 28_123:
      break

    num_list[value-1] = True
```

```python
print([i+1 for i in range(len(num_list)) if not num_list[i]])

total_sum = 0
for i in range(len(num_list)):
  if not num_list[i]:
    total_sum += (i + 1)

total_sum
```
