**16. Power Digit Sum**

 and the sum of its digits is

.

What is the sum of the digits of the number
?

```python
digits_array = [1]

for i in range(1, 1001):
    number_of_digits = len(digits_array)
    count = 0
    remainder = 0

    while count < number_of_digits:
      product = digits_array[count] * 2 + remainder
      digits_array[count] = product % 10
      remainder = product // 10
      count += 1

    while remainder > 0:
      digits_array.append(remainder % 10)
      remainder //= 10

print(sum(digits_array), len(digits_array))
digits_array
```
