**20. Factorial Digit Sum**

image.png

```python
factorial = [1]

for i in range(1, 101):
  remainder = 0

  for j in range(0, len(factorial)):
    multi = factorial[j] * i + remainder
    factorial[j] = multi % 10
    remainder = multi // 10

  while remainder > 0:
    factorial.append(remainder % 10)
    remainder = remainder // 10

# factorial.reverse()
print(sum(factorial))
```
